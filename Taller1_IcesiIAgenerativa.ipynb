Es importante destacar que este proceso se plantea de forma pragmática, evitando generar latencias significativas en los flujos actuales y que se puedan ir cambiando.

También tener un sistemas basado en reglas sería muy limitado y el UX  del usuario no seria el mejor.

La arquitectura propuesta seria:



**Planeador**[ se encarga de dar la bienvenida y recibe la intencion del usuario]

↓


**Verificador** [se encarga de buscar en la base de datos la información dependiendo de la intención del usuario (catalogos, orderlogistic…. ) ]

## Riesgos Éticos:
Este es un punto crucial. ¿Qué riesgos pueden surgir?

- Podria dar informacion erróneo o alucinaciones a los usuarios.

### Sesgo:
El modelo podría reflejar sesgos de los datos de entrenamiento, ofreciendo respuestas preferenciales a ciertos grupos de clientes.  
En este caso resulta fundamental ser muy específicos en la definición de cada uno de los prompts que utilizará el agente, de modo que se eviten inconvenientes o respuestas inadecuadas. Es importante tener en cuenta este aspecto desde el diseño inicial para garantizar mayor precisión y consistencia en las interacciones.

### Privacidad de datos:
¿Cómo se manejaría la información sensible de los clientes (direcciones, historial de compras) si se utiliza para afinar el modelo o como contexto en los prompts?

Uno de los riesgos más relevantes es el manejo de la información sensible de los clientes (direcciones, historial de compras, teléfonos). Para mitigar este riesgo, es fundamental que exista un agente supervisor que valide el contexto y asegure que dichos datos se usen únicamente para los fines autorizados.  
Una alternativa práctica es que, en la API que integre el servicio, en lugar de enviar directamente los datos sensibles, se utilice un número de referencia o identificador único. Con este identificador, el sistema puede recuperar la dirección o el dato correspondiente desde una base segura. De esta forma, si el request contiene un error (por ejemplo, una dirección equivocada o inválida), el envío se cancela automáticamente y se genera una nueva solicitud, evitando exponer información personal innecesariamente.
